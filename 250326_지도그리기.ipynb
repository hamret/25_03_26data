```python
import folium

map = folium.Map(location= [37.53, 126.97],
           zoom_start=15,
           zoom_control=True,
           control_scale=True)
map
```

```python
mk5 = folium.Circle([37.53, 126.97], popup='위치5',
                    Color = 'green')
mk5.add_to(map)

map
```

```python
mk6 = folium.Circle([37.53, 126.97], popup='위치6',
                    Color = 'green', radius=100)
mk6.add_to(map)

map
```

```python
mk7 = folium.Circle([37.53, 126.97], popup='위치6',
                    Color = 'red', radius=300, fill=True)
mk7.add_to(map)

map
```

```python
from folium.plugins import MarkerCluster

map = folium.Map()
folium.Marker([37.53, 126.97]).add_to(map)

mcVar = MarkerCluster().add_to(map)
folium.Marker([37.53, 126.97]).add_to(mcVar)

map
```

```python
import pandas as pd

df = pd.read_excel('/content/자연휴양림출력화일.xlsx')
df
```

```python
vemap = folium.Map(location=[37.538240218883, 127.09689071008],
                   zoom_start=15,
                   zoom_control=False,
                   control_scale= True)
vemap
```

```python
for n, row in df.iterrows():
  print(n, row['휴양림명'], row['위도'], row['경도'])
```

```python
for n, row in df.iterrows():
  vm = folium.Marker(location=[row['위도'],row['경도']], popup = folium.Popup(row['휴양림명'], max_width=200))
  vm.add_to(vemap)

vemap
```

```python
vemap2 = folium.Map(location=[37.538240218883, 127.09689071008],
                    zoom_start=10,
                    zoom_control=False,
                    control_scale=True)
vemap2
```

```python
from folium.plugins import MarkerCluster

veVar = MarkerCluster().add_to(vemap2)

for n, row in df.iterrows():
  folium.Marker(location=[row['위도'],row['경도']], popup = folium.Popup(row['휴양림명'], max_width=200)).add_to(veVar)

vemap2
```
